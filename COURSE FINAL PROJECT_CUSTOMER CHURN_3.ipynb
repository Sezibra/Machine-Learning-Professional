### 3 — Data Cleaning + Feature Engineering

	•	Ensure numeric types
	•	Handle missing values
	•	Separate features/target

```python
# Convert all columns to numeric safely
for c in df.columns:
    df[c] = pd.to_numeric(df[c], errors="coerce")

TARGET = "churn"

X = df.drop(columns=[TARGET]).copy()
y = df[TARGET].astype(int).to_numpy()

# Median imputation WITHOUT sklearn
X_imputed = X.copy()
medians = X_imputed.median(numeric_only=True)
X_imputed = X_imputed.fillna(medians)

# sanity check
print("Missing in X:", int(X_imputed.isna().sum().sum()))
print("Missing in y:", int(pd.isna(y).sum()))
X_imputed.head()
```

### 4 — Train/Validation/Test Split + Scaling

```python
import numpy as np

# X_imputed is a DataFrame, y can be Series/array
X_np = X_imputed.to_numpy(dtype=np.float32)
y_np = np.asarray(y, dtype=int)

def stratified_train_val_test_split(X, y, test_size=0.20, val_size=0.20, seed=42):
    rng = np.random.default_rng(seed)

    idx0 = np.where(y == 0)[0]
    idx1 = np.where(y == 1)[0]
    rng.shuffle(idx0); rng.shuffle(idx1)

    def split_class(idx):
        n = len(idx)
        n_test = int(round(n * test_size))
        n_rem = n - n_test
        n_val = int(round(n_rem * val_size))

        test_idx = idx[:n_test]
        val_idx  = idx[n_test:n_test+n_val]
        train_idx = idx[n_test+n_val:]
        return train_idx, val_idx, test_idx

    tr0, va0, te0 = split_class(idx0)
    tr1, va1, te1 = split_class(idx1)

    train_idx = np.concatenate([tr0, tr1])
    val_idx   = np.concatenate([va0, va1])
    test_idx  = np.concatenate([te0, te1])

    rng.shuffle(train_idx); rng.shuffle(val_idx); rng.shuffle(test_idx)

    X_train, y_train = X[train_idx], y[train_idx]
    X_val,   y_val   = X[val_idx],   y[val_idx]
    X_test,  y_test  = X[test_idx],  y[test_idx]
    return X_train, X_val, X_test, y_train, y_val, y_test

X_train, X_val, X_test, y_train, y_val, y_test = stratified_train_val_test_split(
    X_np, y_np, test_size=0.20, val_size=0.20, seed=42
)

# StandardScaler replacement: fit on train only
eps = 1e-8
mu = X_train.mean(axis=0)
sd = X_train.std(axis=0)

X_train_s = (X_train - mu) / (sd + eps)
X_val_s   = (X_val   - mu) / (sd + eps)
X_test_s  = (X_test  - mu) / (sd + eps)

print("Train/Val/Test:", X_train_s.shape, X_val_s.shape, X_test_s.shape)
print("Churn rate (train):", round(float(y_train.mean()), 3))
print("Churn rate (val):", round(float(y_val.mean()), 3))
print("Churn rate (test):", round(float(y_test.mean()), 3))
```

```python
import os
OUTDIR = "churn_outputs"
os.makedirs(OUTDIR, exist_ok=True)
print("Saving plots to:", os.path.abspath(OUTDIR))
```

```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
import os

def predict_proba(model, X):
    return model.predict(X, verbose=0).reshape(-1)

def confusion_matrix_np(y_true, y_pred):
    y_true = np.asarray(y_true).astype(int)
    y_pred = np.asarray(y_pred).astype(int)
    tn = int(((y_true == 0) & (y_pred == 0)).sum())
    fp = int(((y_true == 0) & (y_pred == 1)).sum())
    fn = int(((y_true == 1) & (y_pred == 0)).sum())
    tp = int(((y_true == 1) & (y_pred == 1)).sum())
    return np.array([[tn, fp], [fn, tp]])

def auc_tf(y_true, y_prob):
    m = tf.keras.metrics.AUC()
    m.update_state(y_true, y_prob)
    return float(m.result().numpy())

def classification_report_np(y_true, y_pred):
    cm = confusion_matrix_np(y_true, y_pred)
    tn, fp = cm[0, 0], cm[0, 1]
    fn, tp = cm[1, 0], cm[1, 1]

    def safe_div(a, b):
        return a / b if b != 0 else 0.0

    prec1 = safe_div(tp, tp + fp)
    rec1  = safe_div(tp, tp + fn)
    f11   = safe_div(2 * prec1 * rec1, prec1 + rec1)

    prec0 = safe_div(tn, tn + fn)
    rec0  = safe_div(tn, tn + fp)
    f10   = safe_div(2 * prec0 * rec0, prec0 + rec0)

    acc = safe_div(tp + tn, tp + tn + fp + fn)

    return (
        "Classification report (numpy)\n"
        "---------------------------------\n"
        f"Class 0 -> precision: {prec0:.3f}  recall: {rec0:.3f}  f1: {f10:.3f}\n"
        f"Class 1 -> precision: {prec1:.3f}  recall: {rec1:.3f}  f1: {f11:.3f}\n"
        f"Accuracy: {acc:.3f}\n"
        "---------------------------------\n"
        f"Confusion matrix [[TN FP],[FN TP]]:\n{cm}\n"
    )

def eval_binary(model, X, y_true, threshold=0.5):
    y_prob = predict_proba(model, X)
    y_pred = (y_prob >= threshold).astype(int)

    y_true = np.asarray(y_true).astype(int)
    acc = float((y_pred == y_true).mean())
    auc = auc_tf(y_true, y_prob)
    cm = confusion_matrix_np(y_true, y_pred)

    return {
        "accuracy": acc,
        "auc": auc,
        "confusion_matrix": cm,
        "report": classification_report_np(y_true, y_pred)
    }

def plot_history(history, name):
    plt.figure()
    plt.plot(history.history.get("loss", []), label="train_loss")
    plt.plot(history.history.get("val_loss", []), label="val_loss")
    plt.title(f"Loss: {name}")
    plt.xlabel("epoch")
    plt.ylabel("loss")
    plt.legend()
    plt.tight_layout()
    plt.savefig(os.path.join(OUTDIR, f"loss_{name}.png"), dpi=160)
    plt.show()
```

### 7 — Model Variation 1: Baseline MLP

```python
import pandas as pd
import requests
from io import StringIO

url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/" \
      "IBMDeveloperSkillsNetwork-ML0101EN-SkillsNetwork/labs/Module%203/data/ChurnData.csv"

response = requests.get(url, verify=False)
df = pd.read_csv(StringIO(response.text))

print("df shape:", df.shape)
df.head()
```

```python
import numpy as np

TARGET = "churn"

# ensure numeric
for c in df.columns:
    df[c] = pd.to_numeric(df[c], errors="coerce")

X = df.drop(columns=[TARGET])
y = df[TARGET].astype(int).to_numpy()

X_imputed = X.fillna(X.median(numeric_only=True))

print("X_imputed shape:", X_imputed.shape)
print("y shape:", y.shape)
print("Missing values:", X_imputed.isna().sum().sum())
print("Churn rate:", round(y.mean(), 3))
```

```python
X_np = X_imputed.to_numpy(dtype=np.float32)
y_np = y.copy()

def stratified_split(X, y, test_size=0.2, val_size=0.2, seed=42):
    rng = np.random.default_rng(seed)

    idx0 = np.where(y == 0)[0]
    idx1 = np.where(y == 1)[0]
    rng.shuffle(idx0); rng.shuffle(idx1)

    def split(idx):
        n = len(idx)
        n_test = int(n * test_size)
        n_val = int((n - n_test) * val_size)
        return idx[n_test+n_val:], idx[n_test:n_test+n_val], idx[:n_test]

    tr0, va0, te0 = split(idx0)
    tr1, va1, te1 = split(idx1)

    train_idx = np.concatenate([tr0, tr1])
    val_idx   = np.concatenate([va0, va1])
    test_idx  = np.concatenate([te0, te1])

    rng.shuffle(train_idx)
    rng.shuffle(val_idx)
    rng.shuffle(test_idx)

    return (
        X[train_idx], X[val_idx], X[test_idx],
        y[train_idx], y[val_idx], y[test_idx]
    )

X_train, X_val, X_test, y_train, y_val, y_test = stratified_split(X_np, y_np)

# scale
eps = 1e-8
mu = X_train.mean(axis=0)
sd = X_train.std(axis=0)

X_train_s = (X_train - mu) / (sd + eps)
X_val_s   = (X_val   - mu) / (sd + eps)
X_test_s  = (X_test  - mu) / (sd + eps)

print("Shapes:", X_train_s.shape, X_val_s.shape, X_test_s.shape)
print("Train churn rate:", round(y_train.mean(), 3))
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

tf.random.set_seed(42)

model_v1 = keras.Sequential([
    layers.Input(shape=(X_train_s.shape[1],)),
    layers.Dense(32, activation="relu"),
    layers.Dense(16, activation="relu"),
    layers.Dense(1, activation="sigmoid")
])

model_v1.compile(
    optimizer=keras.optimizers.Adam(1e-3),
    loss="binary_crossentropy",
    metrics=[keras.metrics.AUC(name="auc"), "accuracy"]
)

history_v1 = model_v1.fit(
    X_train_s,
    y_train,
    validation_data=(X_val_s, y_val),
    epochs=20,
    batch_size=32,
    verbose=1
)
```

```python
plot_history(history_v1, "v1_baseline")
```

```python
results_v1 = eval_binary(model_v1, X_test_s, y_test)

print("Accuracy:", round(results_v1["accuracy"], 3))
print("AUC:", round(results_v1["auc"], 3))
print(results_v1["report"])
```

### 9 — Model Variation 3: BatchNorm + Class Weighting (handles imbalance better)

```python
import numpy as np

# y_train is already a NumPy array
neg = (y_train == 0).sum()
pos = (y_train == 1).sum()
total = neg + pos

class_weight = {
    0: total / (2 * neg),
    1: total / (2 * pos)
}

print("Class weights:", class_weight)
```

```python
from tensorflow import keras
from tensorflow.keras import layers

model_v3 = keras.Sequential([
    layers.Input(shape=(X_train_s.shape[1],)),
    layers.Dense(64),
    layers.BatchNormalization(),
    layers.Activation("relu"),
    layers.Dropout(0.25),

    layers.Dense(32),
    layers.BatchNormalization(),
    layers.Activation("relu"),
    layers.Dropout(0.15),

    layers.Dense(1, activation="sigmoid")
])

model_v3.compile(
    optimizer=keras.optimizers.Adam(1e-3),
    loss="binary_crossentropy",
    metrics=[keras.metrics.AUC(name="auc"), "accuracy"]
)

model_v3.summary()
```

```python
hist_v3 = model_v3.fit(
    X_train_s,
    y_train,
    validation_data=(X_val_s, y_val),
    epochs=200,
    batch_size=32,
    class_weight=class_weight,
    callbacks=callbacks,
    verbose=0
)

plot_history(hist_v3, "v3_class_weighted")

v3_val = eval_binary(model_v3, X_val_s, y_val)
v3_test = eval_binary(model_v3, X_test_s, y_test)

v3_val, v3_test["auc"]
```

Model Performance Summary: 

	•	Model 1 (Baseline MLP) achieved strong overall accuracy but demonstrated limited ability to identify churners. While it performed well in predicting non-churn customers, its low churn recall indicates that many at-risk customers would remain undetected.

	•	Model 2 (Dropout + L2 Regularization) delivered the strongest overall performance.
